**Notebook: Analysis of Generative Art Dataset**

This notebook performs a detailed analysis of a dataset of generative art pieces, exploring various models and styles, and their frequencies.
Each section of the notebook will delve into:
- Data loading and initial exploration
- Analysis of generative models and styles used
- Visualization of findings

### Import Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Load the dataset

```python
# read metadata.csv
metadata = pd.read_csv('metadata.csv')
```

 Display basic information about the dataset

```python
metadata.info()
metadata.head()
```

```python
# Standardization: put all data in style, period and models columns in lowercase
metadata['style'] = metadata['style'].str.lower()
metadata['period'] = metadata['period'].str.lower()
metadata['generative_model'] = metadata['generative_model'].str.lower()
```

```python
# changing in style column simbolism to symbolism
metadata['style'] = metadata['style'].str.replace('simbolism', 'symbolism')
```

## Total Paintings
Here we calculate and print the total number of paintings in the dataset.

```python
total_paintings = metadata.shape[0]
print(f'Total number of paintings: {total_paintings}')
```

# Count of Paintings per Generative Model
We analyze and visualize the frequency of each generative model in the dataset.


```python
model_counts = metadata['generative_model'].value_counts()
model_counts.plot(kind='bar')
plt.title(f'Number of paintings per generative model, total: {total_paintings}')
plt.xlabel('Generative model')
plt.ylabel('Number of paintings')
plt.yticks(np.arange(0, max(model_counts)+1, 1))
plt.show()
```

```python
# In percentage
model_counts_percentage = model_counts / total_paintings * 100
for model, count in model_counts_percentage.items():
    print(f'{model}: {count:.2f}%')
```

# Count of Paintings per Style
Now, we analyze the distribution of art styles in the dataset.

```python
style_counts = metadata['style'].value_counts()
style_counts.plot(kind='bar', color='blue')
plt.title(f'Number of paintings per style, total: {total_paintings}')
plt.xlabel('Style')
plt.ylabel('Number of paintings')
plt.show()
```

```python
# In percentage
style_counts_percentage = style_counts / total_paintings * 100
for style, count in style_counts_percentage.items():
    print(f'{style}: {count:.2f}%')
```
